### Graphical Exploration - Distributions

```python
sns.set(rc = {'figure.figsize': (14, 10)})
sns.lineplot(x = 'cut', y = 'total_sales_price', ci = None, data = diamonds)
plt.show()
```

Results: This shows that the cushion cut diamond is the most expensive and round cut is the least expensive.

```python
sns.set(rc = {'figure.figsize': (14, 10)})
sns.lineplot(x = 'cut_quality', y = 'carat_weight', ci = None, data = diamonds)
plt.show()
```

Results: This shows that most people purchase a middle of the road 'Good' quality diamond because they are able to get a higher carat weight.

```python
sns.set(rc = {'figure.figsize': (14, 10)})
sns.lineplot(x = 'lab', y = 'total_sales_price', ci = None, data = diamonds)
plt.show()
```

Results: This shows that the HRD certified diamonds are more expensive and make more money than IGI and GIA. This standard is more common than IGI and GIA. GIA, or natural diamonds, make slightly more money and are a little more expensive than IGI, or lab grown diamonds.

### Graphical Exploration - Scatter Plots

```python
plt.scatter(diamonds['table_percent'], 
           diamonds['carat_weight'])
plt.xlabel('Table Percent')
plt.ylabel('Carat Weight')
plt.title('Table Percent vs Carat Weight')
plt.grid(True)
plt.show()
```

Results: The average table percent of a diamond is between 45 and 80 percent which is average for a diamond.

```python
plt.scatter(diamonds['meas_length'], 
           diamonds['carat_weight'])
plt.xlabel('Meas Length')
plt.ylabel('Carat Weight')
plt.title('Meas Length vs Carat Weight')
plt.grid(True)
plt.show()
```

Results: This shows that the meas length grows exponentially as the carat weight grows bigger, the length also gets bigger. There are a few outliers, but the main cluster shows that these are positively correlated between each other.

### Graphical Exploration - Pie Plot

```python
diamonds['color'].value_counts().plot(kind = 'pie')
plt.show()
```

Results: The pie plot shows that there are more people buying coloress diamonds, G, D, F, E, than those with a little color, like L, M, J, and K.

### Graphical Exploration - Bar Plots

```python
diamonds['cut_quality'].value_counts().plot(kind = 'barh')
plt.show()
```

Results: This graph shows that the better cut to a diamond, the more expensive it will be.

```python
sns.barplot(x = 'color', y = 'total_sales_price', data = diamonds)
plt.show()
```

Results: This graph shows that the nicer the color the higher the cost will be for that specific diamond.

### Graphical Exploration - Other Plots

```python
diamond_values = diamonds[(diamonds['color']=='G') | (diamonds['color']=='D') | 
                          (diamonds['color']=='F') | (diamonds['color']=='E')]
```

```python
sns.set(rc = {'figure.figsize': (10,9)})
sns.lineplot(x = 'carat_weight', y = 'total_sales_price', 
             hue = 'color', ci = None, style = 'color', data = diamond_values)
plt.show()
```

Results: This shows that the higher the carat weight the more a diamond will cost.

```python
sns.set(rc = {'figure.figsize': (10,9)})
sns.barplot(x = 'color', y = 'carat_weight', ci = None, data = diamonds)
plt.show()
```

Results: This bar plot shows that the diamonds with the highest carat weight are the ones that are in the middle of the color categories where they have a little color, but aren't colorless or completely yellow. The best option for the highest carat weight would be colors in the middle like L and M.

```python
sns.set(rc = {'figure.figsize': (10,9)})
sns.lineplot(x = 'color', y = 'total_sales_price', data = diamonds)
plt.show()
```

Results: This goes with the previous graph and shows that a person can buy a middle of the road diamond like L and M for a fraction of the cost with more carat weight. This graph shows that the better color to a diamond the more it will cost, but the previous graph also states that the better the color the less carat weight a person can get. These show that the L and M colors have the best price for the best carat weight.

```python
sns.set(rc = {'figure.figsize': (10,9)})
sns.barplot(x = 'cut_quality', y = 'carat_weight', ci = None, data = diamonds)
plt.show()
```

Results: This shows that a person can buy a middle of the road diamond like L and M for a fraction of the cost with more carat weight and an average cut. This means that the better cut to a diamond, the more it will cost, but the previous graph also states that the better the color the less carat weight a person can get. These show that the L and M colors have the best price for the best carat weight. Along with this, an average cut allows for a person to have a bigger carat weight for a better price.
